# Analysis

```python
# Imports
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sqlite3
import os
```

```python
# Database connection
db_path = os.path.join('..', 'db', 'tickets.db')
conn = sqlite3.connect(db_path)
print("Connection Successfull", db_path)
```

```python
# Load data into DataFrame
df = pd.read_sql_query("SELECT * FROM tickets", conn)
df
```

```python
conn.close()
```

```python
print("\nShape:", df.shape)
df.info()
display(df.head())
```

```python
# Check missing values
missing = df.isnull().sum().sort_values(ascending=False)
missing = missing[missing > 0]
print("\nMissing Values:")
print(missing)
```

```python
# Basic statistics
display(df.describe(include='all'))
```

```python
#  Date columns to datetime
for col in ['creationdate', 'lastupdated', 'deadline', 'closuretime']:
    if col in df.columns:
        df[col] = pd.to_datetime(df[col], errors='coerce')
```

```python
# Target variable analysis
if 'state' in df.columns:
    plt.figure(figsize=(10, 5))
    sns.countplot(data=df, x='state', order=df['state'].value_counts().index)
    plt.title("Ticket State Distribution")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

```python
# Time to closure
if 'creationdate' in df.columns and 'closuretime' in df.columns:
    df['time_to_close'] = (df['closuretime'] - df['creationdate']).dt.total_seconds() / 3600
    df['time_to_close'].hist(bins=30, figsize=(10, 5))
    plt.title("Time to Close (Hours)")
    plt.xlabel("Hours")
    plt.ylabel("Ticket Count")
    plt.show()
```

```python
# Monthly trend (if yearmonth available)
if 'yearmonth' in df.columns:
    df['yearmonth'] = df['yearmonth'].astype(str)
    df['yearmonth'].value_counts().sort_index().plot(kind='bar', figsize=(12, 5))
    plt.title("Tickets Created Per Month")
    plt.xticks(rotation=90)
    plt.show()
```

```python
# Ageing Bucket
if 'ageing_bucket' in df.columns:
    sns.countplot(data=df, x='ageing_bucket', order=df['ageing_bucket'].value_counts().index)
    plt.title("Ticket Ageing Buckets")
    plt.show()
```

```python
#  Assigned Agent
if 'assignedagent' in df.columns:
    top_agents = df['assignedagent'].value_counts().head(10)
    top_agents.plot(kind='barh', title='Top 10 Assigned Agents', figsize=(10, 5))
    plt.xlabel("Ticket Count")
    plt.show()
```

```python
# Correlation Heatmap (numeric only)
numeric_df = df.select_dtypes(include='number')
plt.figure(figsize=(10, 6))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()
```

```python
# Save cleaned version (optional)
out_path = os.path.join('..', 'data', 'processed', 'tickets_cleaned.csv')
df.to_csv(out_path, index=False)
print(f"\nCleaned data saved to: {out_path}")
```
